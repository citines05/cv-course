```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
import torch
import torchvision
import ultralytics
```

```python
image = cv2.imread("street.png")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# Obj detection

model = ultralytics.YOLO("yolo11n.pt")

results = model(image)

for i, result in enumerate(results):
    result.save(filename=f"results{i}.jpg")

results_image = cv2.imread("results0.jpg")
plt.imshow(cv2.cvtColor(results_image, cv2.COLOR_BGR2RGB))
```
