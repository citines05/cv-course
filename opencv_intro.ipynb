```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('image.png')
print(type(image))
plt.imshow(image)
```

```python
# primeiro tranformar a imagem para float
gain = np.clip(image / 255 * 1.001, 0, 1) # limita o vetor resultante a espaços entre 0 e 1
plt.imshow(gain)
```

```python
# convertendo para o espaço de cores correto
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gain_rgb = np.clip(image_rgb / 255 * 1.001, 0, 1)
plt.imshow(gain_rgb)
```

```python
crop = image_rgb[0:100, 0:100] # posição inicial e final nos eixos x e y
plt.imshow(crop)
```

```python
larger = cv2.resize(image_rgb, (int(image.shape[1] * 4), int(image.shape[0] * 4))) # multiplica a resolução da imagem por 4
plt.imshow(larger)
```
