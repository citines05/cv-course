```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('image.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# operação de corte de imagem
crop = image[0:100, 0:100]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
```

```python
# Dimensões da imagem og (translação)
height, width = image.shape[:2]

# Matriz de translação
translation_x, translation_y = 8, 4
translation_matrix = np.float32([
    [1, 0, translation_x],
    [0, 1, translation_y]
])

# aplicando a matriz na transformação da imagem
translated = cv2.warpAffine(image, translation_matrix, (width, height))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
```

```python
# Dimensões da imagem original (rotação)
height, width = image.shape[:2]

# definindo o centro da rotação
center = (width // 2, height // 2)

# define o ângulo de rotação
angle = 45

# define uma operação de escala da imagem
scale = 1.0

# deriva a matriz de rotação
rottation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# aplica a matriz de transformação da imagem
rotated = cv2.warpAffine(image, rottation_matrix, (width, height))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
```

```python
image_draw = cv2.rectangle(image, (10, 20), (50, 50), (0, 0, 255), 5) # pt1 ponto mais alto à esquerda e pt2 ponto mais baixo à direita
plt.imshow(cv2.cvtColor(image_draw, cv2.COLOR_BGR2RGB))
```
